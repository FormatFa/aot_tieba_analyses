```python
#当前年 
nowYear = '2019'
#转换数据为日期类型,now为这个json爬取的日期,替换时间为日期时要用
def to_date(createtime,now):
    #用-分隔
    temp = createtime.split("-")
    #时间的的没有-，所以长度是1
    if len(temp)==1:
        return now
    #年形式的,返回空
    elif len(temp[0])==4:
        return np.nan
    #月-日形式的，这里个1,2..这些前面补个0
    else:
        return nowYear+'-'+'-'.join(temp)
```

```python

#读取单个json,返回dataframe
def parsejson(path):
    print("解析文件:",path)
    df = pd.read_json(path,lines=True)
    df = df.drop_duplicates(subset=['tid'])
    #根据文件名获取当前日期
    name=os.path.split(path)[1].replace(".json","")
    print("文件名日期:",name)
    #处理日期
    df['createdate']=df['createtime'].apply(to_date,args=(name,))
    #格式化成日期类型
    df['createdate']=pd.to_datetime(df['createdate'],format="%Y-%m-%d")
    return df
```

```python
jsons = os.listdir(jsondir)
dataframes=[]
for json in jsons:
    dataframes.append(parsejson(os.path.join(jsondir,json)))
threads = pd.concat(dataframes)
```

```python
#需要作图的月,最近三个月的数据比较靠谱
months=[8,9,10]
#y轴日期,虽然不是每个月都有31天
dates=[i for i in range(1,32)]

print(dates)
#各月的y轴数据

#提取的数据，每个月(key) ,对应的帖子数列表
values = {}
for i in months:
        #计算这个月有多少天
        start=pd.Timestamp(year=int(nowYear),month=i,day=1)
        #不确定结束是多少号，这里获取下一个月的一号，减去一天就是
        end = pd.Timestamp(year=int(nowYear),month=i+1,day=1)-pd.Timedelta(days=1)
        #对应的帖子数
        temp = day_titles[start:end].values
        counts = []
        #有些天数不是31天的，需要把列表补全
        for j in range(0,31):
            if j < len(temp):
                counts.append(temp[j])
            else:
                counts.append(0)
        print(i) 
        values[i]=counts
```

各个月的分词词频统计
- 统计各个月的标题的词

```python
#加载需要的名词
import jieba
jieba.load_userdict('aot_dict.txt')
```

```python
#加载停用词表
stops = [ i.strip() for i in  open("stopword.txt","r").read().split("\n")]
```

```python
#将每个月的所有标题合成一个字符串，再分词统计
```

```python
month_titles=threads[['createdate','title']].groupby(threads['createdate'].apply(lambda x:x.month)).agg(lambda item:",".join(item))
```

```python
#计算词频,输入为分词后的结果
def wc(words):
    result={}
    for word in words:
        #过滤最大的两个出来
        if word=="艾伦" or word=="巨人":
            continue
        #过滤停用词
        if word in stops or word.strip()=='':
            continue
        #第一次没有
        if word not in result:
            result[word]=1
        else:
            result[word]+=1
    return result

```

```python
#排序 取出前100个
result={}
#month_titles.index
for i in months:
    words = wc( jieba.cut(month_titles.loc[i].title)  )
    words = sorted( words.items(),key=lambda x:x[1],reverse=True)[0:100]
    result[i]=words
```

```python
#词云相关
from wordcloud import WordCloud
```

```python
#转换成字典
for m in months:
    wld = WordCloud(width=512,height=512,background_color="white")
    wcs = { i[0]:i[1] for i in result[m]}
    #参数要是一个字典
    wld.generate_from_frequencies(wcs)
    wld.to_file(str(m)+"月.png")
```
